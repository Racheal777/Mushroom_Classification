# DATA PREPROCESSING

```python
#Using Label Encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
mushroom = mushroom.apply(LabelEncoder().fit_transform)

mushroom.head()
```

```python
#Assigning the dependent to y
x = mushroom.drop(["class"], axis = 1)
y = mushroom["class"]
```

```python
#Spliting the data set to train and test data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
mushroom = DecisionTreeClassifier(random_state = 0, max_depth = 5)
mushroom.fit(x_train, y_train)
```

```python
mushroom.score(x_train, y_train)
```

```python
#Predicting the x_test
predictions = mushroom.predict(x_test)
```

```python
#Predicting the accuracy of the y_test and the x_test which has been predicted in predictions from above

from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

# Random Forest Model

```python
#Importing the Random Forest Classifier Library

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth =5)
```

```python
# Predicting the score

rf.fit(x_train, y_train)
rf.score(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
```

```python
rf.score(x_test, y_test)
```

```python
accuracy_score(y_test, predictions)
```

Decision Tree Accuracy  = 98%

Random Forest Accuracy = 99%

Hence Random forest is Comparatively better
